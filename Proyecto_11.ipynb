Construya un objeto en python que permita llevar a cabo un ACP. El objeto debe entregar la proyecci√≥n de los datos sobre todos los componentes principales en un dataframe y la matriz de inercia vista en clase.

```python
import numpy as np
import pandas as pd
import openpyxl as open
```

```python
class PCA:

  def __init__(self,X,scal=True):
    self.X=X
    self.scal=scal
    self.vec=None
  
  def inercia(self):
    if self.scal==True:
      df2_scal = np.zeros(X.shape)
      for i in range(df2_scal.shape[1]):
        df2_scal[:,i] = (X.iloc[:,i]-X.iloc[:,i].mean())/X.iloc[:,i].std()
      self.X=df2_scal
    else:
      self.X=self.X

    S = self.X.T.dot(self.X)/self.X.shape[0]
    eigvals, self.vec = np.linalg.eig(S)
    varianza_explicada = {pos:i/sum(eigvals) for pos, i in enumerate(eigvals)}
    inercia = pd.DataFrame({'eigvals':eigvals,
                       'inercia_acum':np.cumsum(eigvals),
                       'v_expl':varianza_explicada.values(),
                       'v_expl_acum':np.cumsum([i for i in varianza_explicada.values()])})
    return(inercia.T)

  
  def proyeccion(self):
    Names = ['0']*X.shape[1]
    for i in range(1,X.shape[1] + 1):
      Names[i-1] = 'ACP' + str(i)
    proyeccion= self.X.dot(self.vec[:,:self.X.shape[1]])
    proyeccion= pd.DataFrame(proyeccion, columns = Names)

    return(proyeccion)
```
